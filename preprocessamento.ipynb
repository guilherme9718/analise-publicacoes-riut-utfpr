```python
import pandas as pd
```

```python
data = []
with open('dados/articles.json', 'r') as f:
    data = f.readlines()

import json
data_dict = []
for text in data:
    try:
        d = json.loads(text[0:-2])
        data_dict.append(d)
    except:
        pass

data_dict[0:3]
df = pd.DataFrame.from_dict(data_dict)
df
```

## Remover duplicados

```python
df = df.loc[df['title'].drop_duplicates().index]
df
```
