```python
import pandas as pd
```

```python
data = []
with open('dados/articles.json', 'r') as f:
    data = f.readlines()

import json
data_dict = []
for text in data:
    try:
        d = json.loads(text[0:-2])
        data_dict.append(d)
    except:
        pass

data_dict[0:3]
df = pd.DataFrame.from_dict(data_dict)
df
```

## Remover duplicados

```python
df = df.loc[df['title'].drop_duplicates().index]
df
```

## Tratar a data

```python
df.loc[0, 'date'] = '11-Fev-2015'
```

```python
from datetime import datetime, date
import locale
locale.setlocale(locale.LC_ALL, "pt_BR.utf8")
formats = ("%d-%b-%Y", "%b-%Y", "%d-%b-%y", "%Y")

def parse_date(date):
    exs = []
    for fmt in formats:
         try:
            return datetime.strptime(date.replace(" ", ""), fmt)
         except Exception as ex:
            exs.append(ex)
    raise exs[0]
df['date'] = df['date'].apply(parse_date)
df['date']
```

```python
df = df.drop(['advisor2'], axis=1)
df
```

```python
df['unique_id'] = df['uri'].str.extract(r"(/1)(/\d+)")[1].str.replace("/", "").astype(int)
df = df.set_index('unique_id')
```
